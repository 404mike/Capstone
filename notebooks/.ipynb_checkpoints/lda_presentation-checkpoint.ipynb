```python
#some helper functions for vector addition, subtraction, and comparision
def vec_lst_add_subtract(top_vec_1, top_vec_2, mode='add'):
    '''
    Utility to add and subtract vectors in the format of a list of tuples
    '''
    res_vec = []
    
    if mode == 'add':
        for pr1, pr2 in zip(top_vec_1, top_vec_2):
            sm = pr1[1] + pr2[1]
            if sm > 1:
                sm = 1
            res_vec.append((pr1[0], sm))
    elif mode == 'subtract':
        for pr1, pr2 in zip(top_vec_1, top_vec_2):
            diff = pr1[1] - pr2[1]
            if diff < 0:
                diff = .0000000000000000001
            res_vec.append((pr1[0], diff))
     #I THINK I NEED TO ADD RENORMALIZATION OF THE RESULTANT!!       
    
    else:
        print("Invalid mode")
        return 0
    
    return res_vec


def find_most_similar(sim_vec, all_top_vecs, title_lst, vec_in_corp='Y', n_results=7):                
    '''
    Calculates cosine similarity across the entire corpus and returns 
    the n_results number of most similar documents
    '''
    
    cos_sims = [gensim.matutils.cossim(sim_vec, vec) for vec in all_top_vecs]
    
    if vec_in_corp == 'N':
        most_similar_ind = np.argsort(cos_sims)[::-1][:n_results]
    if vec_in_corp == 'Y':
        most_similar_ind = np.argsort(cos_sims)[::-1][:n_results+1][1:]
        #exclude 'self', in the case that it is a book in the corpus
    
    for ind in most_similar_ind:
        print (title_lst[ind], cos_sims[ind])
        
        
def find_topic_books(topic_num, all_top_vecs, n_results=10):
    '''
    Finds the books that have the strongest match to the selected topic number
    '''
    one_topic_lst = [vec[topic_num][1] for vec in all_top_vecs]
    return np.argsort(one_topic_lst)[::-1][:n_results]
    
```

```python
#helper function to display the manually assigned topic designations for a document's topic vector
def get_topic_desig(doc_topics_lst, topic_dict=topic_dict):
    '''
    Translate the numerical topics in a '.get_document_topics' list into the assigned topic designation 
    '''
    descriptive_topic_lst = []
    
    for pr in doc_topics_lst:
        if pr[0] in topic_dict: 
            descriptive_topic_lst.append((topic_dict[pr[0]], pr[1]))
        else: 
            descriptive_topic_lst.append(('unassigned_topic', pr[1]))
    
    return descriptive_topic_lst
        
```

## Section 5: Document Similarity

Here we will compare books in the corpus using cosine similarity to compare the topic vector representations of these documents.

```python
url = 'http://cs.carleton.edu/cs_comps/0910/netflixprize/final_results/knn/img/knn/cos.png'
Image(url,width=300, height=500)
```

The 'magic' that makes this all work is that our lda model has allowed us to represent each book as a mixture of n different topics, where n is the number of topics that the model is trained on.  If our topic model is good, then books that have similar topic distributions of the same topics should be intuitively similar to each other!  

_**In effect we have reduced the number of dimensions required to represent an *entire book* to around 50 or less!!**_

Grab all the document topic vectors and stick them in a list that we can use later:

```python
#WARNING: Long execution time - 20-30 mins
if 0 == go_no_go:
    all_top_vecs = [lda.get_document_topics(serial_corp[n], minimum_probability=0) \
                    for n in range(len(serial_corp))]
```

See below for an example of one of these vectors.

_Note we ensure that the vector has a length of 50 by setting 'minimum probability=0'.  This is important because we want to be comparing same-length vectors for this implementation of addition/subtraction/similarity._

```python
pandp_top_vec = all_top_vecs[bks.pandp[0]]

print(bks.pandp)
pandp_top_vec
```

**Book Similarity**

```python
find_most_similar(pandp_top_vec, all_top_vecs, title_lst, \
                  vec_in_corp='Y', n_results=11)
```

_**Whoa, we identified two other Jane Austen books as similar to Pride & Prejudice!!  That is cool!!**_   

Next let's find out what our model thinks is similar to the King James Bible:

```python
bible_top_vec = all_top_vecs[bks.kjv_bible[0]]

find_most_similar(bible_top_vec, all_top_vecs, title_lst, vec_in_corp='Y')
```

## Section 6: Book Algebra!

Since we have vector representations of our documents, we can also add and subtract books, and then find the books in the corpus that are most similar to the resulting sum or difference!

By adding and subtracting our n-dimensional book topic vectors, we generate new document topic vectors, that encode a different mixture of topics.  We can then find the book most similar to the 'sum' or 'difference' of two books by comparing that new resultant vector to the rest of the document topic vectors in the corpus via cosine similarity.

```python
diff_vec = vec_lst_add_subtract(pandp_top_vec, bible_top_vec, mode='subtract')

find_most_similar(diff_vec, all_top_vecs, title_lst, vec_in_corp='N')
```

Notice that these are basically the same results as the similarity for the Pride & Prejudice itself.  This checks out, because the KJV Bible is essentially just one topic, the 'religious scriptures' topic.  Pride & Prejudice does not include this topic, so we can't really take it out.  

Let's try to sum them next:

```python
sum_vec = vec_lst_add_subtract(pandp_top_vec, bible_top_vec, mode='add')
find_most_similar(sum_vec, all_top_vecs, title_lst, vec_in_corp='N')
```

Here our resultant vector is quite different!  It's an interesting mix of Shakespeare, other 18th/19th-century literature, and one particularly interesting result: The Torn Bible.  This book is about a priest falling in love with young girl, so we can see the new coexistance of topics of religion and romantic love.

Next we'll look at the classic 20,000 Leagues Under the Sea and a book called The Life of a Ship, which was selected for its high correlation with the 'ship' topic:

```python
life_ship_top_vec = all_top_vecs[bks.life_ship[0]]
twentyk_leagues_top_vec = all_top_vecs[bks.twentyk_leagues[0]]

diff_vec2 = vec_lst_add_subtract(twentyk_leagues_top_vec, life_ship_top_vec, mode='subtract')
find_most_similar(diff_vec2, all_top_vecs, title_lst, vec_in_corp='N')
```

The books that are most similar to this resultant vector all appear to relate to nature and biology, but notably absent are any titles that relate to the ocean or ships!

Finally, a little dark fun...

What do you get when you add Lewis Carroll's fantastical nonsense children's story with Joseph Conrad's dark and gritty store of imperialism and the worst of human nature?  Let's find out...

```python
alice_top_vec = all_top_vecs[bks.alice[0]]
hod_top_vec = all_top_vecs[bks.heart_darkness[0]]

sum_vec2 = vec_lst_add_subtract(hod_top_vec, alice_top_vec, mode='add')
find_most_similar(sum_vec2, all_top_vecs, title_lst, vec_in_corp='N')
```

In this list I see books about supernatural occurrances in the wilderness, and dark and mystical themes.  Sounds like a good combination of the two to me!
